````python
import pandas as pd
import matplotlib.pyplot as plt

# Extract the year from 'QUOTE_DATE' and create a new column 'QUOTE_YEAR'
df['QUOTE_YEAR'] = df['QUOTE_DATE'].dt.year

# Group the data by 'QUOTE_YEAR' and 'STRIKE', then count the occurrences
strike_counts_by_year = df.groupby(['QUOTE_YEAR', 'STRIKE']).size().reset_index(name='COUNT')

# Get unique years from the data
years = strike_counts_by_year['QUOTE_YEAR'].unique()

# Create a grouped bar chart for strike prices and counts for each year
plt.figure(figsize=(12, 6))  # Set the figure size

# Iterate through years and plot strike prices and counts
for year in years:
    data_year = strike_counts_by_year[strike_counts_by_year['QUOTE_YEAR'] == year]
    plt.bar(data_year['STRIKE'], data_year['COUNT'], label=str(year), alpha=0.7)

# Customize the plot
plt.title('Strike Prices and Counts for Each Year')
plt.xlabel('Strike Price')
plt.ylabel('Count')
plt.legend(title='Year')

# Show the plot
plt.grid()
plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Extract the year from 'QUOTE_DATE' and create a new column 'QUOTE_YEAR'
df['QUOTE_YEAR'] = df['QUOTE_DATE'].dt.year

# Group the data by 'QUOTE_YEAR' and 'STRIKE', then count the occurrences
strike_counts_by_year = df.groupby(['QUOTE_YEAR', 'STRIKE']).size().reset_index(name='COUNT')

# Get unique years from the data
years = strike_counts_by_year['QUOTE_YEAR'].unique()

# Set the number of columns for subplots
num_cols = 3  # You can adjust the number of columns as needed
num_rows = (len(years) + num_cols - 1) // num_cols

# Create subplots with Seaborn style
plt.figure(figsize=(15, 10))

for i, year in enumerate(years):
    data_year = strike_counts_by_year[strike_counts_by_year['QUOTE_YEAR'] == year]
    plt.subplot(num_rows, num_cols, i + 1)
    sns.barplot(data=data_year, x='STRIKE', y='COUNT', palette='viridis', alpha=0.8)
    plt.title(f'Year {year}', fontsize=14)
    plt.xlabel('Strike Price', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()

# Remove any empty subplots
if len(years) < num_cols * num_rows:
    for i in range(len(years), num_cols * num_rows):
        plt.delaxes(plt.subplot(num_rows, num_cols, i + 1))

# Add a main title
plt.suptitle('Strike Prices and Counts for Each Year', fontsize=16, y=1.02)

# Show the plot
plt.tight_layout()
plt.show()
````

# Explore Expiry Date

````python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded your DataFrame 'df'

# Extract contract expiry dates
expiry_dates = df['EXPIRE_DATE']

# Plot a histogram of contract expiry dates
plt.figure(figsize=(10, 6))
plt.hist(expiry_dates, bins=30, edgecolor='k')
plt.xlabel('Contract Expiry Dates')
plt.ylabel('Count')
plt.title('Distribution of Contract Expiry Dates')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with option contract data

# Convert 'EXPIRE_DATE' to a pandas datetime object if it's not already
df['EXPIRE_DATE'] = pd.to_datetime(df['EXPIRE_DATE'])

# Filter the DataFrame for contracts in the year 2019
year_2019_contracts = df[df['EXPIRE_DATE'].dt.year == 2019]

# Group the contracts by 'STRIKE' and calculate the mean of 'C_LAST' for each strike price
top_10_contracts = year_2019_contracts.groupby('STRIKE')['C_LAST'].mean().nlargest(10)

# Plot the 10 best contracts by strike price and their average 'C_LAST'
plt.figure(figsize=(12, 6))
top_10_contracts.plot(kind='bar', edgecolor='k')
plt.title('Top 10 Option Contracts by Average C_LAST Price (Year 2019)')
plt.xlabel('Strike Price')
plt.ylabel('Average C_LAST Price')
plt.xticks(rotation=45)
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Assuming 'df' is your DataFrame with option contract data

# Choose a random contract from the DataFrame
random_contract = df.sample(1)

# Extract the contract's 'STRIKE' and 'EXPIRE_DATE' values
strike_price = random_contract['STRIKE'].values[0]
expire_date = random_contract['EXPIRE_DATE'].values[0]

# Filter the DataFrame for the selected contract
selected_contract = df[(df['STRIKE'] == strike_price) & (df['EXPIRE_DATE'] == expire_date)]

# Sort the DataFrame by 'QUOTE_DATE' for a time series plot
selected_contract = selected_contract.sort_values(by='QUOTE_DATE')

# Create a line plot of 'QUOTE_DATE' against 'C_LAST'
plt.figure(figsize=(12, 6))
plt.plot(selected_contract['QUOTE_DATE'], selected_contract['C_LAST'], marker='o', linestyle='-', markersize=3)
plt.title(f'Option Contract (Strike Price: {strike_price}, Expiry Date: {expire_date}) Over Time')
plt.xlabel('QUOTE_DATE')
plt.ylabel('C_LAST Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
````
